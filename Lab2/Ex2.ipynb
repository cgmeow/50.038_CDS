```python
import pandas as pd
terms1 = pd.Series(['this', 'is', 'a', 'sample'])
term_count1 = pd.Series([1,1,2,1])
terms2 = pd.Series(['this','is','another','document'])
term_count2 = pd.Series([1,1,2,3])

df1 = pd.DataFrame({ 'Term': terms1, 'Frequency': term_count1 })
df2 = pd.DataFrame({ 'Term': terms2, 'Frequency': term_count2 })
df2
```

```python
import math
def tfidf(w,d):
    max_frequency = d['Frequency'].max()
    word_frequency = d.loc[d['Term']==w,'Frequency'].iloc[0]
    tf = word_frequency/max_frequency
    
    count =0
    if w in terms1.values:
        count +=1

    if w in terms2.values:
        count +=1
        
    idf = math.log(2/count,2)
    tfidf_value = tf*idf
    return tf, idf


tfidf('document', df2)
```
